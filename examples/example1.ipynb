## Usuage

To use WQChartPy, you have to parpare the geochemical data into a csv file. I parpared a data templete in the data folder. It is something like below.

### Load data

```python
import pandas as pd

# Load the data
df = pd.read_csv('../data/data_template.csv')

# Show the data
df
```

1.	The first column named Sample stores the sample names or IDs. 

2.	The second column named Label indicates which group a sample belongs to. For example, the C1, C2, C3, and C4 indicate the samples are from four clusters. This column is very important as it is used to distinguish samples from different groups in the plots.

3.	The third column named Color is used to store the colors shown for the samples. A complete list of colors can be found here https://matplotlib.org/stable/gallery/color/named_colors.html.
	
4.	The fourth column named Marker is used to store the marker styles shown for the samples. A complete list of markers can be found here https://matplotlib.org/stable/api/markers_api.html.
	
5.	The fifth column named Size is used to control the values of marker size shown for the samples. 
	
6.	The sixth column named Alpha is used to change the opacity of the markers. The alpha value should be between 0 and 1.

7.	The remaining columns store the concentration values of the geochemical parameters. All the concentrations should be in mg/L except for pH in standard unit.

Note that the column names cannot be changed. However, the orders of these columns are not required to be same as shown above.  

### Gibbs plot

```python
from wqchartpy.plot import gibbs

gibbs.plot(df, unit='mg/L', figname='Gibbs diagram', figformat='jpg')
```

Note there are four argeuments for the plot_Gibbs function. The df is the DataFrame by reading the data by Pandas. The unit tells the function all the concentration values are in mg/L. Currently, mg/L is the only supported unit. The figname and figformat are used to save the generated figure to local disk.

The four argeuments are commonly used in wqchartpy,  as shown below.

### Piper diagram - triangle

```python
from wqchartpy.plot import triangle_piper

triangle_piper.plot(df, unit='mg/L', figname='trianglePiper diagram', figformat='jpg')
```

### Piper diagram - rectangle

```python
from wqchartpy.plot import rectangle_piper

rectangle_piper.plot(df, unit='mg/L', figname='rectanglePiper diagram', figformat='jpg')
```

### Piper diagram - color coded

```python
from wqchartpy.plot import color_piper

rgb = color_piper.plot(df, unit='mg/L', figname='color coded Piper diagram', figformat='jpg')
```

You may note the except for the genereated figure, color_piper.plot_Piper also returns the RGB values in the two trangles and the central diamond.

### Durvo diagram

```python
from wqchartpy.plot import durvo

durvo.plot(df, unit='mg/L', figname='Durvo  diagram', figformat='jpg')
```

### HFE-D diagram

```python
from wqchartpy.plot import hfed

hfed.plot(df, unit='mg/L', figname='HFE-D diagram', figformat='jpg')
```

### Stiff diagram

```python
from wqchartpy.plot import stiff

stiff.plot(df, unit='mg/L', figname='Stiff diagram', figformat='jpg')
```

The stiff.plot_Stiff would plot the stiff diagram for each sample.
